```python
import sqlite3
import numpy as np
import matplotlib.pylab as plt
import pandas as pd
import seaborn as sns
pd.set_option('display.max_rows', 200)
pd.set_option('display.max_columns', 200)
```

```python
con = sqlite3.connect("employee_analysis.sqlite3")
df = pd.read_sql("SELECT * FROM employees", con)
```

### Employment Tenure (Histogram joined with Density flow)

```python
hist = sns.displot(df, x="Employment Tenure", hue="CompanyName", kde=True, binwidth=10).set(title="Endava employment tenure", ylabel="No. employees", xlabel="Employment tenure (months)")
hist.savefig("endava-emp-ten.png", transparent=True)
```
